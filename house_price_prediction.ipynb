Import Necessary Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

Load Dataset kc_house_data.csv

```python
data = pd.read_csv('kc_house_data.csv')
```

Explore Dataset

```python
data.head()
```

Data Visualisation - Plotting the Distribution of House Prices

```python
plt.hist(data['price'], bins=50)
plt.xlabel('Price')
plt.ylabel('Number of houses')
plt.title('Distribution of house prices in King County, USA')
plt.show()
```

Plotting the Correlation Between House Price and Other Features

```python
corr = data.corr()
sns.heatmap(corr, annot=True)
plt.title('Correlation matrix of house price and other features')
plt.show()
```

Data Preparation - Splitting data 

```python
x = data[['bedrooms','bathrooms']]
y = data['price']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

Creating Linear Regression Model

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

Make Predictions on the Testing Set

```python
y_pred = model.predict(x_test)
```

Evaluate the Model Performance

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print('Mean squared error:', mse)
print('R-squared:', r2)
```
